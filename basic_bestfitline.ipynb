```python
plt.scatter(x,y)
```

```python
m=7
b=100
z=m*x
z=z+b
plt.plot(x,z,color='red')
```

```python
def cost(p):
  c=0
  for i in range(30):
    c+=(y[i]-p[i])*(y[i]-p[i])

  return c
```

```python
print("Initial Cost:", cost(z))
```

```python
def update(m,b,lr,lr2):
  #old implementation
  #grad=0
  #gradb=0
  #for i in range(30):
    #grad+=(y[i]-m*x[i]-b)*-1*x[i]
    #gradb+=-(y[i]-m*x[i]-b)

  grad=np.sum(np.dot(-1*(x.T),y-(m*x + b)))
  gradb=np.sum(-1*(y-(m*x + b)))

  M=m-grad*lr
  B=b-gradb*lr2
  return M,B
```

```python
print("Checking Update Function",update(5,100,0.000001,0.0001))
```

```python
print("Improving Loss:")
```

```python
m=-2
b=-200
z=m*x
z=z+b
for i in range(300):
  if i%10==0:
    print(cost(z))
  m,b=update(m,b,0.000001,0.001)
  z=m*x
  z=z+b
```
